```python
plt.figure(figsize=(8,4))
sns.countplot(x='Sales Channel', hue='Response', data=file1)
plt.ylabel('number of responses')
plt.title('Responses by Sales Channel')
plt.show()
```

```python
file1.boxplot('Total Claim Amount')
```

```python
sns.histplot(data=file1, x="Total Claim Amount", hue="Response", multiple="stack")
```

```python
sns.boxplot(x="Income", y="Response", data=file1)
```

```python
sns.histplot(data=file1, x="Income", hue="Response", multiple="stack")
```
